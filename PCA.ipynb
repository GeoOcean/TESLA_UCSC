# PCA

In this notebook, a simple application of Principal Component Analysis (PCA) is presented. <br>
The notebook comprises two simple examples in **2D** and **1D** dimensions.

```python
import warnings
warnings.simplefilter(action='ignore', category=DeprecationWarning)

#bluemath path
import sys
import os
```

```python
from bluemath_tk.datamining.pca_v2 import PCA
from bluemath_tk.datasets import DATA
```

## 2D Example

### Step 1. Initialize PCA class

```python
pca_ob = PCA()
```

**Set Data**

```python
ds = DATA.get_2d_dataset()
```

### Step 2. Setup PCA parameters

```python
pca_ob.dataset = ds
pca_ob.dim_pca = 'coord3' # coordinate to do PCA over
pca_ob.generate_data_matrix()
```

```python
pca_ob.plot_data()
```

### Step 3. PCA Analysis


```python
pca_ob.pca()
```

#### Plot PCA results

```python
#pca_ob.plot_eofs()
```

```python
#pca_ob.plot_pcs()
```

```python
pca_ob.plot_eofs_pcs(n_pca = 3)
```

## 1D Example

### Step 1. Initialize PCA class

```python
pca_ob1 = PCA()
```

**Set Data**

```python
ds1 = ds.isel(coord2=0).squeeze().drop('coord2').rename({'coord3':'coord2'})
```

### Step 2. Setup PCA parameters

```python
pca_ob1.dataset = ds1
pca_ob1.dim_pca = 'coord2' # coordinate to do PCA over
pca_ob1.generate_data_matrix()
```

```python
cp = {
            'figsize': (len(pca_ob.vars_keys)*8, 10),
        }

pca_ob1.plot_data(custom_params = cp)
```

### Step 3. PCA Analysis


```python
pca_ob1.pca()
```

#### Plot PCA results

```python
#pca_ob1.plot_eofs()
```

```python
#pca_ob1.plot_pcs()
```

```python
pca_ob1.plot_eofs_pcs(n_pca = 3)
```
