## Data

For KMA class usage, a pandas dataframe is required. Each column will represent a different variable

```python
df = DATA.get_1d_dataframe()
df
```

## KMA

```python
KMA()
```

```python
# Step 1. Initialize KMA class

kma_ob = KMA()

# Step 2. Setup data and scalar and directional variables

kma_ob.data = df.values
kma_ob.ix_scalar = [0, 1, 2]  # Scalar columns
kma_ob.ix_directional = [] #Directional columns 
```

```python

# Step 3. Define number of cluster for KMA

M = 10
kma_ob.kma(M)
```
